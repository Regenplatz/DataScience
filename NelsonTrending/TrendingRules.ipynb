```python
import pandas as pd
import numpy as np
```

```python
def calculateMean(data):
    return np.mean(data)

def calculateSigma(data):
    return np.std(data)

def simulateDataInGivenRange():
    ## numbers between -1 and 1 with mean = 0
    data = np.arange(-1, 1, 0.001)
    np.random.shuffle(data)
    print("MEAN:", np.mean(data), "\nSTD :", np.std(data))
    print("MIN :", np.min(data), "\nMAX :", np.max(data))
    return data
```

```python
## simulate 1000 data points (normally distributed): exemplarily demonstrated
data = np.arange(-1, 1, 0.001)
print("MEAN:", np.mean(data), "\nSTD :", np.std(data))
print("MIN :", np.min(data), "\nMAX :", np.max(data))
```

## Rule 01: One point is more than 3 standard deviations from the mean.
Problem indicated: One sample is grossly out of control.

```python
## append 1 data point that is more than 3 std from the mean
data = simulateDataInGivenRange()
rule01_data = np.append(data, 4)
print("\nMEAN (with added sample(s)):", np.mean(rule01_data), 
      "\nSTD  (with added sample(s)):", np.std(rule01_data))
```

```python
mean = calculateMean(rule01_data)
sigma = calculateSigma(rule01_data)

for elem in rule01_data:
    if elem < (mean - 3*sigma) or elem > (mean + 3*sigma):
        print("outlier")
```

## Rule 02: Nine (or more) points in a row are on the same side of the mean.
Problem indicated: Some prolonged bias exists.

```python
rule02_data = [1.4, -0.6, 0.9, 2, 2.5, 1.5, 1.04, 1.05, 1.2, 1.04, 1.06, -1.9, -0.3, 1.5] 

mean = calculateMean(rule02_data)
sigma = calculateSigma(rule02_data)
# print("Mean:", mean, "\nSigma:", sigma, "\n")

sign = []
for elem in rule02_data:
    if elem < mean:
        sign.append("neg")        
        if len(sign) >= 9:
            if "pos" not in sign[-9:]:
                print("outlier\n", sign)
    else:
        sign.append("pos")
        if len(sign) >= 9:
            if "neg" not in sign[-9:]:
                print("outlier\n", sign)
#     print(sign)    
```

## Rule 03: Six (or more) points in a row are continually increasing (or decreasing).
Problem indicated: A trend exists.

```python
rule03_data = [1.4, 1.04, 1.06, -1.9, 1.6, 1.9, 2, 2.5, 3.8, 4.2, 5.05, 6.2, 1.04, 1.06, 1.5] 

mean = calculateMean(rule03_data)
sigma = calculateSigma(rule03_data)
print("Mean:", mean, "\nSigma:", sigma, "\n")
```

```python
## initialize with first value
trend_increasing = [rule03_data[0]]
trend_decreasing = [rule03_data[0]]


for elem in rule03_data[1:]:
    
    ## continually increasing
    if len(trend_increasing) > 0:
        if elem > trend_increasing[-1]:
            trend_increasing.append(elem)
    else:
        trend_increasing = [elem]
    
    ## continually decreasing
    if len(trend_increasing) > 0:
        if elem < trend_decreasing[-1]:
            trend_decreasing.append(elem)
    else:
        trend_decreasing = [elem]
        
    ## output if 6 in a row
    if len(trend_increasing) >= 6:
        print("Increasing Trend!")
    elif len(trend_decreasing) >= 6:
        print("Decreasing Trend!")
```

## Rule 04: Fourteen (or more) points in a row alternate in direction, increasing then decreasing.
Problem indicated: This much oscillation is beyond noise. Note that the rule is concerned with directionality only. <br>
The position of the mean and the size of the standard deviation have no bearing.

```python
rule04_data = [1.4, 1.04, 1.06, -1.9, 1.6, 1.2, 2, -2.5, 3.8, -4.2, 5.05, -6.2, 1.04, -1.06, 1.5] 

mean = calculateMean(rule04_data)
sigma = calculateSigma(rule04_data)
print("Mean:", mean, "\nSigma:", sigma, "\n")
```

```python
## initialize with first value
trend_alternating = [rule04_data[0]]

alternating = [rule04_data[0]]
sign = [""]
for elem in rule04_data[1:]:
    
    ## alternating in- and decreasing
    if len(trend_alternating) > 0:
        
        if elem > alternating[-1]:
            alternating.append(elem)
            if len(sign) > 0 and sign[-1] != "pos":
                sign.append("pos")
            else:
                trend_alternating = [elem] 
        else:
            alternating.append(elem)
            if len(sign) > 0 and sign[-1] != "neg":
                sign.append("neg")
            else:
                trend_alternating = [elem] 
    else:
        trend_alternating = [elem]
        
#     print(alternating, "\n", sign, "\n\n")
    
    ## output if 14 alternating in a row
    if len(alternating[1:]) >= 14:
        print("Alternating Trend!")
```

## Rule 05: Two (or three) out of three points in a row are more than 2 standard deviations from the mean in the same direction.
Problem indicated: There is a medium tendency for samples to be mediumly out of control.<br>
The side of the mean for the third point is unspecified.

```python
## append 1 data point that is more than 3 std from the mean
data = simulateDataInGivenRange()
rule05_data = np.append(data, [1.6, 1.5, 1.5])

print("\nMEAN (with added sample(s)):", np.mean(rule05_data), 
      "\nSTD  (with added sample(s)):", np.std(rule05_data))
```

```python
mean = np.mean(rule05_data) 
sigma = np.std(rule05_data)


## if element is greater than the mean    
for idx, elem in enumerate(rule05_data):
    last3 = list(rule05_data[:idx+1])[-3:]
    if len(last3) == 3:
        if min(last3) > mean:
            min2 = last3.copy()
            min2.remove(min(last3))
            if min(min2) > mean + 2*sigma:
                print("outlier : ", last3)

## if element is smaller than the mean    
for idx, elem in enumerate(rule05_data):
    last3 = list(rule05_data[:idx+1])[-3:]
    if len(last3) == 3:
        if min(last3) < mean:
            min2 = last3.copy()
            min2.remove(min(last3))
            if min(min2) < mean - 2*sigma:
                print("outlier : ", last3)
 
```

## Rule 06: Four (or five) out of five points in a row are more than 1 standard deviation from the mean in the same direction.
Problem indicated: There is a strong tendency for samples to be slightly out of control.<br>
The side of the mean for the fifth point is unspecified.

```python
## append 1 data point that is more than 1 std from the mean
data = simulateDataInGivenRange()
rule06_data = np.append(data, [1.6, 1.5, 1.5, 0.9, 2.0, 1.2])

print("\nMEAN (with added sample(s)):", np.mean(rule06_data), 
      "\nSTD  (with added sample(s)):", np.std(rule06_data))
```

```python
mean = np.mean(rule06_data) 
sigma = np.std(rule06_data)


## if element is greater than the mean    
for idx, elem in enumerate(rule06_data):
    last5 = list(rule06_data[:idx+1])[-5:]
    if len(last5) == 5:
        if min(last5) > mean:
            min4 = last5.copy()
            min4.remove(min(last5))
#             ## check on > 2*sigma is easier for principle check as check on > 1*sigma
#             if min(min4) > mean + 2*sigma:
            if min(min4) > mean + sigma:
                print("outlier : ", last5)

## if element is smaller than the mean    
for idx, elem in enumerate(rule06_data):
    last5 = list(rule06_data[:idx+1])[-5:]
    if len(last5) == 5:
        if min(last5) < mean:
            min4 = last5.copy()
            min4.remove(min(last5))
#             ## check on > 2*sigma is easier for principle check as check on > 1*sigma
#             if min(min4) < mean - 2*sigma:
            if min(min4) < mean - sigma:
                print("outlier : ", last5)
 
```

## Rule 07: Fifteen points in a row are all within 1 standard deviation of the mean on either side of the mean.
Problem indicated: With 1 standard deviation, greater variation would be expected.

```python
## append 1 data point that is more than 1 std from the mean
data = simulateDataInGivenRange()
rule07_data = np.append(data, [0.1, 0.2, 0.3, -0.1, -0.2, -0.3, 0.1, 0.2, 0.3, -0.1, -0.2, -0.3, -0.2, -0.1, 0.2, 0.1])

print("\nMEAN (with added sample(s)):", np.mean(rule07_data), 
      "\nSTD  (with added sample(s)):", np.std(rule07_data))
```

```python
mean = np.mean(rule07_data) 
sigma = np.std(rule07_data)
    
    
## if 15 elements in a row are within 1 std of the mean    
for idx, elem in enumerate(rule07_data):
    last15 = list(rule07_data[:idx+1])[-15:]
    if len(last15) == 15:
        if min(last15) > (mean - sigma) and max(last15) < (mean + sigma):
            print("outlier : ", last15)
```

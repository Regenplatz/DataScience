## ELT PROCESS
To demonstrate NoSQL skills, the ELT process was widened although not being necessary. This means that for this project two load and transform steps were implemented.  

In particular, the **Extract-Transform-Load** process is performed as follows:
1. `Extract`: Retrieve data from different sources. 
2. `Load`: Load raw data to the Mongo database *Space*
3. `Transform`: Transform the data by applying aggregation pipelines. Select fields of interest and perform transformations such as calculations if needed. 
4. `Load`: Load transformed data to another MongoDB (called *Outside_Earth*).
5. `Transform`: Analyze the transformed data by applying aggregation pipelines.

### Dataflow / System architecture
This project focused mainly on our solarsystem, a few supernovae and the location of the International Space Station (ISS). <br>

For this, raw data was fetched as a first step and loaded to MongoDB. Raw data was then processed within this jupyter notebook by applying various aggregations. In the following, transformed data was loaded to another MongoDB. This step was in principle not needed, but demonstrates various options. Based on those created collections *solarsystem*, *supernovae* and *iss*, analysis were performed. For the analysis of the ISS location, a reference collection (which contained information on several cities with their corresponding coordinates and country) was fetched and loaded to MongoDB.

### Fetch Data

#### Data concerning *Solar System* and *ISS*

```python
## fetch data related to iss whereabouts and solar system
def fetchData(api):
    """
    Fetch data from url via API and load to JSON
    :param api: String, API by which data is accessed
    :return: dictionary, retreived data
    """
    ## define name for print statement
    if "systeme-solaire" in api:
        name = "Solar System"
    elif "satellites" in api: 
        name = "ISS"
  
    ## request data from website and convert to json
    request_data = requests.get(api)
    data = json.loads(request_data.text)  
    
    print(f"\n{name} ------------------------\n", 
          request_data.text[0:450])
    
    return data
    

data_solsys = fetchData(API_URL_SOLARSYSTEM)
data_iss = fetchData(API_URL_ISS)
```

### Transform and load to MongoDB
##### Transform collection *solarsystem*

```python
## show structure of an exemplary document
pprint(raw_solarsystem.find_one({}))
```

<P style="page-break-after:always;">

<br>


```python
## build query "solarsystem" and write to MongoDB
solarsystem = raw_solarsystem.aggregate([
    
    ## create new fields for VALUE * 10^EXPONENT
    {"$addFields": {"vol [km^3]": {"$multiply": ["$vol.volValue", 
                                            {"$pow": [10, "$vol.volExponent"]}]},
                    "mass [kg]": {"$multiply": ["$mass.massValue", 
                                            {"$pow": [10, "$mass.massExponent"]}]} }},
                    
    ## self left join collection to access the document where "aroundPlanet.planet"  
    ## equals "id" in another document. This is done because the "aroundPlanet.planet" 
    ## field reveals the French planet name but the English name is wanted.    
    {"$lookup": {"from": "raw_solarsystem",
                 "localField": "aroundPlanet.planet",
                 "foreignField": "id",
                 "as": "aroundPlanet(oid)"}},
    
    ## extract name (of planet/planetoid being surrounded) from array 
    {"$unwind": {"path": "$aroundPlanet(oid)", "preserveNullAndEmptyArrays": True}},     

    ## define attributes being displayed
    {"$project": {"_id": 0,
                  "name": "$englishName",
                  "isPlanet": 1,
                  "bodyType": 1,
                  "sideralOrbit [d]": "$sideralOrbit",
                  "moons": 1,
                  "gravity [m/s^2]": "$gravity",
                  "mass [kg]": 1,
                  "vol [km^3]": 1, 
                  "density [g/cm^3]": "$density",                  
                  "sideralRotation [h]": "$sideralRotation",
                  "avgTemp [K]": "$avgTemp",
                  "aroundPlanet(oid)": "$aroundPlanet(oid).englishName" }},
    
    ## write to database "Outside_Earth" as collection "solarsystem"
    {"$out": {"db": "Outside_Earth", "coll": "solarsystem"}}
])
```

```python
## load collection 
solsys = client.Outside_Earth["solarsystem"]

## display in dataframe, omit MongoDB _id's and set name as index
df_solarsystem = pd.DataFrame(list(solsys.find()))
df_solarsystem.drop("_id", axis=1, inplace=True)
df_solarsystem.set_index("name", inplace=True)
df_solarsystem.head(5)
```

<P style="page-break-after:always;">

<br>


```python
## in case mass and density are given, calculate volume=mass/(density*10^12).
## write calculated value into new field (so you can compare in case the 
## volume is already given)
for document in solsys.find({}):     
    if document["mass [kg]"] != None and document["density [g/cm^3]"] != None:    
        solsys.update_one({"_id": document["_id"]}, 
                           {"$set": {'vol_calculated': 
                                     document["mass [kg]"] / 
                                     (document["density [g/cm^3]"] * 10**12)}
                          })
```

```python
## show data types for all fields in one document
for doc in solsys.find():    
    for key in doc.keys():        
        print(key, type(doc[key]))
    break
```

```python
## show all field names in a document
solsys_fields = solsys.aggregate([
    {"$project": {"arrayofkeyvalue": {"$objectToArray":"$$ROOT"}}}, 
    {"$unwind": "$arrayofkeyvalue"}, 
    {"$group":{"_id":None, "allkeys":{"$addToSet": "$arrayofkeyvalue.k"}}} ])

for record in solsys_fields: 
    print(record["allkeys"])
```

##### Transform collection *supernovae*

```python
## build query "supernovae" and display in dataframe
supernovae = raw_supernovae.aggregate([
    
    ## create new fiels
    {"$addFields": {
        "redshift": "$redshift.value",
        "velocity [km/s]": "$velocity.value",
        "photometry_time": "$photometry.time",
        "photometry_magnitude": "$photometry.magnitude"}},
    
    ## extract from arrays 
    {"$unwind": "$redshift"},
    {"$unwind": "$velocity [km/s]"},
    {"$unwind": "$photometry_time"},
    {"$unwind": "$photometry_magnitude"},
    
    ## define attributes being displayed and convert some to double format
    {"$project": {"_id":0,
        "name": 1,
        "redshift": {"$toDouble": "$redshift"},
        "velocity [km/s]": {"$toDouble": "$velocity [km/s]"},
        "photometry_time": {"$toDouble": "$photometry_time"},
        "photometry_magnitude": {"$toDouble": "$photometry_magnitude"}}},
    
    ## write to database "Outside_Earth" as collection "supernovae"
    {"$out": {"db": "Outside_Earth", "coll": "supernovae"}}
])
```

```python
## load collection 
supnov = client.Outside_Earth["supernovae"]

## display in dataframe, omit MongoDB _id's and set name as index
df_supernovae = pd.DataFrame(list(supnov.find()))
df_supernovae.drop("_id", axis=1, inplace=True)
df_supernovae.set_index("name", inplace=True)
df_supernovae.head(5)
```

##### Transform collection *iss*
Create new fields for coordinates, UTC and CET datetimes. Select fields of interest and load to MongoDB "Outside_Earth". 

```python
## show structure of the only document in this raw data collection
pprint(raw_iss.find_one({}))
```

```python
iss = raw_iss.aggregate([    
    ## create field for [longitude, latitude] as well as for date time 
    ## (extracted from timestamp)
    {"$addFields": {"coordinates [lng/lat]": ["$longitude", "$latitude"],
                    "date_time (UTC)": {"$toDate": {"$toLong": 
                                            {"$multiply": ["$timestamp", 1000]}}} }},
    
    ## create new field for date time as CET time. Default was earlier 
    ## provided as UTC time.
    {"$addFields": {"date_time (CET)": {"$dateToString": {"date": "$date_time (UTC)",
                                            "format": "%Y-%m-%d %H:%M:%S",
                                            "timezone": "Europe/Berlin" }} }},
    
    ## select fields of interest                  
    {"$project":{"name": 1,
                 "coordinates [lng/lat]": 1,
                 "_id": 0,
                 "velocity [km/h]": "$velocity",
                 "visibility": 1,
                 "date_time (UTC)": 1,
                 "date_time (CET)": 1 }},
    
    ## write to database "Outside_Earth" as collection "iss"
    {"$out": {"db": "Outside_Earth", "coll": "iss"}}
])
```

```python
## display gravity (descending order) of planets/planetoids within a dataframe
q1 = solsys.aggregate([   
    ## select fields of interest
    {"$project":{"name": 1, 
                 "gravity [m/s^2]": 1,
                 "_id": 0}},
    
    ## sort "gravity [m/s^2]" in descending order
    {"$sort": {"gravity [m/s^2]": -1}},
    
    ## limit to the first five results
    {"$limit": 5}
])

df_q1 = pd.DataFrame(q1)
df_q1.set_index("name")
```

<P style="page-break-after:always;">

<br>


```python
## display number of moons orbiting planets/planetoids (within a dataframe)
q2 = solsys.aggregate([
    ## select all moons --> "aroundPlanet(oid)" field not null
    {"$match": {"aroundPlanet(oid)": {"$ne": None}}},
    
    ## select fields of interest
    {"$project":{"name": 1,
                 "aroundPlanet(oid)": 1,
                 "_id": 0}},
    
    ## group by aroundPlanet(oid) according to the number of moons
    {"$group": {"_id": {"aroundPlanet(oid)": "$aroundPlanet(oid)"},
                "cnt_moons": {"$sum": 1}}},
    
    ## select fields of interest
    {"$project":{"_id": 0,
                 "aroundPlanet(oid)":"$_id.aroundPlanet(oid)",
                 "cnt_moons": 1}},    
    
    ## sort "cnt_moons" in descending order
    {"$sort": {"cnt_moons": -1}},
    
    ## limit to the first five results
    {"$limit": 5}
])

df_q2 = pd.DataFrame(q2)
df_q2.set_index("aroundPlanet(oid)")
```

###  Analyze collection *iss*
To evaluate the location of the ISS at the given time of initial raw data loading, reference data is fetched and loaded to MongoDB *Outside_Earth*. This step is generally performed similar to those for the raw data on *solarsystem*, *supernovae* and *ISS*. However, it contains an additional processing step before loading the data into MongoDB. Such a big radius was chosen to higher the chance for obtaining results in case the ISS is located somewhere above the middle of an ocean. Relative distances based on these results are then calculated and sorted in ascending order to get information on the nearest cities. <br>

REMARK: In this approach, earth curvature and altitude were not taken into consideration!

#### Create a reference collection *geolocation* for geocoordinates

```python
## load geodata of various cities to create a reference for iss location
## (above which city is the iss located at the given timestamp?)

## define API, database client and emtpy collection before adding new data
url = "https://raw.githubusercontent.com/lutangar/cities.json/master/cities.json"
API_URL_GEOLOCATION = url
geolocation = client.Outside_Earth["geolocation"]
geolocation.drop()

## request data and load to json
request_data = requests.get(API_URL_GEOLOCATION)
data_geo = json.loads(request_data.text)

## create the field "coord" for the combining longitude and latitude in a list
geoloc_list = []
for doc in data_geo:
    geo_dict = {}
    geo_dict["country"] = doc["country"]
    geo_dict["city"] = doc["name"]
    geo_dict["coord"] = {"coordinates": [float(doc["lng"]), float(doc["lat"])], 
                         "type": "Point"}

    geoloc_list.append(geo_dict)
    
## load data to MongoDB
geolocation.insert_many(geoloc_list)

print("Geolocation ------------------------\n", 
          request_data.text[0:450].replace("\n", "").replace(" ", ""))
```

#### Find the clostest city related to the ISS

```python
## load geo data from MongoDB "Outside_Earth"
geoloc = client.Outside_Earth["geolocation"]

## create index for geospatial data
geoloc.create_index([("coord", "2dsphere")])

## extract ISS coordinates
iss_location = iss_infos.find_one({})
iss_coord = iss_location["coordinates [lng/lat]"]
print("iss coordinates: [", round(iss_coord[0], 5), round(iss_coord[1], 5), "]")

## find the cities that are closely located towards the ISS (in a 2000 km radius)
geo_location = geoloc.find(
   {"coord":
       {"$near": {"$geometry": {
                       "type": "Point",
                       "coordinates": iss_coord},
                  "$maxDistance": 2000000}}
})

## initialize dictionary and list for city coordinates
city_dict = {}
city_dist = []

## extract the closest cities
for elem in geo_location:
    city_coord = elem["coord"]["coordinates"]  
    
    ## calculate relative distance: 
    ## pythagoras, circumference (earth): 40.075 km, 360
    ## NOTE: Earth curvature and altitude is not taken into consideration!
    distance = ( abs(city_coord[0] - iss_coord[0])**2 + 
                 abs(city_coord[1] - iss_coord[1])**2 )**0.5 
    distance = distance / 360 * 40075
    
    ## fill dictionary
    city_dict[elem["city"]] = {"country": elem["country"], 
                               "city_coord": city_coord, 
                               "dist_to_iss": round(distance, 1)}    
    
qx = pd.DataFrame(city_dict).transpose().sort_values("dist_to_iss").head(10)
qx.head(5)
```

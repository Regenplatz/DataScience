```python
import pandas as pd
import numpy as np
import math
from matplotlib import pyplot as plt
import seaborn as sns
from psychrochart import PsychroChart
import psychrolib
import sys
import warnings
warnings.filterwarnings("ignore")
```

## Load Data

```python
df = pd.read_csv("weather_data.csv", sep=",")
df = df[["temperature", "humidity"]].dropna()
df.head()
```

```python
## map temperature and corresponding humidity value to tuple (for each data point)
d = {}
for idx, row in df.iterrows():
#    d[idx] = {"xy": (row["temperature"], row["humidity"])}
    
#scatter_style={"s": 2, "alpha": 0.3}
    
    d[idx] = {"xy": (row["temperature"], row["humidity"]),
          "style": {"markersize": 2, "alpha": 0.3}}
```

## Create Mollier Diagram

```python
## Load preconfigured styles:
print("** Preconfigured chart style: `ashrae`:")

plt.figure(figsize = (12,12))
chart_ashrae_style = PsychroChart('ashrae')
chart_ashrae_style.plot(ax=plt.gca())
chart_ashrae_style.plot_points_dbt_rh(points=d)
```
